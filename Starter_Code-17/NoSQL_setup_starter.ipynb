5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
establishments.update_many({}, [{'$set': {'geocode.longitude': {'$toDouble': '$geocode.longitude'}, 
                                         'geocode.latitude': {'$toDouble': '$geocode.latitude'}
                                         }
                                }
                               ]
                          )
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
query = {"RatingValue": {"$nin": ["1", "2", "3", "4", "5"]}}
update_data = {"$set": {"RatingValue": None}}

result = establishments.update_many(query, update_data)
print(result.modified_count)
```

```python
# Change the data type from String to Integer for RatingValue

query = {
    "RatingValue": {
        "$exists": True,
        "$ne": None
    }
}

documents_to_update = establishments.find(query)

# Iterate through the documents and update RatingValue to integer
for document in documents_to_update:
    new_rating_value = int(document["RatingValue"])
    establishments.update_one(
        {"_id": document["_id"]},
        {"$set": {"RatingValue": new_rating_value}}
    )

# Count the updated documents
updated_count = establishments.count_documents(query)


# Print the count of updated documents
print(f"{updated_count} documents updated.")


```

```python
# Fetch a single document that has 'RatingValue', 'latitude', and 'longitude'
sample_document = establishments.find_one(
    {'RatingValue': {'$exists': True}, 'geocode.latitude': {'$exists': True}, 'geocode.longitude': {'$exists': True}}
)

# Check the data type of 'RatingValue', 'latitude', and 'longitude'
if sample_document:
    rating_value = sample_document.get('RatingValue', None)
    latitude = sample_document.get('geocode', {}).get('latitude', None)
    longitude = sample_document.get('geocode', {}).get('longitude', None)

    print(f"RatingValue is of type: {type(rating_value)}")
    print(f"Latitude is of type: {type(latitude)}")
    print(f"Longitude is of type: {type(longitude)}")
else:
    print("No suitable sample document found.")
```
